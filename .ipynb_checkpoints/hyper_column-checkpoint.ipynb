```python
from matplotlib import pyplot as plt
import numpy as np
#
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Flatten, Dense, Dropout
from tensorflow.python.keras.layers import Conv2D, MaxPooling2D
from tensorflow.python.keras.optimizers import SGD

from tensorflow.python.keras import backend as K

#
from skimage.transform import resize
import skimage
print(skimage.__version__)

#
from sklearn.manifold import TSNE
from sklearn import manifold, cluster
from sklearn.preprocessing import StandardScaler
```

```python
from tensorflow.python.keras.applications import vgg16

#
model = vgg16.VGG16()
sgd=SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy')

# 
#model.summary()
```

```python
import cv2

# last dimension is BGR
# while for matplotlib, last dimension is RGB
im_original=cv2.resize(cv2.imread('dumb.jpg'), (224, 224))  
print(im_original.shape)
im = im_original[:, :, ::-1]  #RGB
plt.imshow(im)
```

```python
im=np.expand_dims(im, axis=0)
out = model.predict(im)
plt.plot(out.reshape(-1))
```

```python
# extracting arbitrary feature maps
layer_index = -5
print(model.layers[layer_index].name)
get_features=K.function([model.layers[0].input], [model.layers[layer_index].output])
#
feat = get_features([im])[0]
print(feat.shape)
len_pics=3
for i in range(len_pics):
    plt.subplot(1, len_pics, i+1)
    feat0 = feat[0, :, :, i]
    upscaled = resize(feat0, (224, 224), preserve_range=True)
    plt.imshow(upscaled[:, :])
    #plt.imshow(feat0)
    plt.xticks([])
    plt.yticks([])
```

```python
# extract hypercolumns
def extract_hypercolumn(model, layer_indices, img):
    layers = [model.layers[i].output if isinstance(i, int) else model.get_layer(i).output \
               for i in layer_indices]
    #
    get_features=K.function([model.layers[0].input], layers)
    feature_maps = get_features([img])  # [nouts, nsamples, height, width, nchs]
    #
    hypercolumns=[]
    for conv_map in feature_maps:
        for fmap in conv_map:
            upscaled = resize(fmap, (224, 224), preserve_range=True)
            hypercolumns.append(upscaled)
            
    return np.concatenate(hypercolumns, axis=2)
```

```python
#layers_extract=[6, 10, 14, -5] = ['block2_pool', 'block3_pool', 'block3_pool', 'block5_pool']
layers_extract=['block2_pool', 'block3_pool', 14, 'block5_pool']

#
hc = extract_hypercolumn(model, layers_extract, im)

# 
print("shape of hypercolumn:{}".format(hc.shape))

#
feat_index = 2
#
plt.subplot(1, 2, 1)
plt.imshow(hc[:, :, feat_index])
plt.xticks([])
plt.yticks([])
plt.title('feature map #{} of {}'.format(feat_index, hc.shape[2]))

# average hypercolumn
ave = np.average(hc, axis=2)

#
print("shape after average of hypercolumn:{}".format(ave.shape))
#
plt.subplot(1, 2, 2)
plt.imshow(ave)
plt.xticks([])
plt.yticks([])
plt.title('average')
```

```python
#
m = hc.reshape(50176, -1)
kmeans = cluster.KMeans(n_clusters=2, max_iter=300, n_jobs=5, precompute_distances=True)
cluster_labels=kmeans.fit_predict(m)

#
print(cluster_labels.shape)
plt.imshow(cluster_labels.reshape(224, 224), cmap='gray')
```
